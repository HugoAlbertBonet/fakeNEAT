```python
import pandas as pd

df = pd.read_excel("NEGA-Iris.xlsx", index_col = 0)
df["n_gen"] += 1
df["fit_test"] = df["fit_test"].apply(lambda x: float(x[7:-1]))
df["time_per_it"] = df["time"]/df["n_gen"]
```

```python
# plot feature importance using built-in function
from xgboost import XGBRegressor
from xgboost import plot_importance
from matplotlib import pyplot
x = df.drop(columns = ["fit_test", "fitness", "time", "time_per_it", "n_gen"])
y = df["fit_test"]
# fit model no training data
model = XGBRegressor()
model.fit(x, y)
# plot feature importance
plot_importance(model)
pyplot.show()
```

```python
# plot feature importance using built-in function
from xgboost import XGBRegressor
from xgboost import plot_importance
from matplotlib import pyplot
x = df.drop(columns = ["fit_test", "fitness", "time", "time_per_it", "n_gen"])
y = df["time_per_it"]
# fit model no training data
model = XGBRegressor()
model.fit(x, y)
# plot feature importance
plot_importance(model)
pyplot.show()
```

```python
# plot feature importance using built-in function
from xgboost import XGBRegressor
from xgboost import plot_importance
from matplotlib import pyplot
df["fit/timexit"] = df["fit_test"]*df["time_per_it"]
x = df.drop(columns = ["fit_test", "fitness", "time", "time_per_it", "n_gen", "fit/timexit"])
y = df["fit/timexit"]
# fit model no training data
model = XGBRegressor()
model.fit(x, y)
# plot feature importance
plot_importance(model)
pyplot.show()
```
